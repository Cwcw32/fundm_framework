## 1.2 pkl文件的写入与读取

```python
import pickle
#wb是覆盖写，如果需要追加，则为‘ab'
f = open('data.pkl','wb')
#待写入数据
datas = {'name':'Bruce','age':25,'high':175}
#写入
data = pickle.dump(datas,f,-1)
#关闭文件
f.close()
```
